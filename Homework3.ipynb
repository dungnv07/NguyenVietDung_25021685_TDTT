```python
#B4
h = int(input())
print ( ' ' * (h-1) + '*')
for i in range(1, h-1):
    print( ' ' * (h-i-1) + '*' + ' ' * (2*i-1) + '*')
print ( '*' * (2*h-1))
```

```python
#B10
a = int(input())
sum = 0
for i in range(1, a):
    if a%i == 0:
        sum += i
    else:
        continue
if sum == a:
    print('yes')
else:
    print('no')
print(sum)
```

```python
#B12
n = int(input())
S = 0
for i in range(n+1):
    S += i
    if S >= n:
        break
        S -= i
print(i-1)
```
