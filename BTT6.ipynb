```python
#W6A1
a = list(map(int, input().split()))
x = []
for i in a:
    if i not in x:
        x.append(i)
print(x)
```

```python
#W6A2
a = list(map(int, input().split()))
x = 0
b = []
for i in a:
    x += i
    b.append(x)
print(b)
```

```python
#W6A3
a = tuple(map(int, input().split()))
k = int(input())
a1 = a[:k]
a2 = a[k:]
print(a2 + a1)
```

```python
#W6A4
ds = input().split()
d = {}
for pair in ds:
    key, value = pair.split(':')
    if key not in d:
        d[key] = [value]
    else:
        d[key].append(value)
print(d)
```

```python
#W6A5
numl = list(map(int, input().split()))
d = {'positives': [], 'negatives': [], 'zeros': []}
for k in numl:
    if k > 0:
        d['positives'].append(k)
    elif k < 0:
        d['negatives'].append(k)
    else:
        d['zeros'].append(k)
print(d)
```

```python
#W6A7
t = tuple(input().split())
l = []
m = 1
for i in t:
    l.append(t[-m])
    m += 1
t2 = tuple(l)
print(t[0], t[-1], t2)
```

```python
#W6A10
numl = list(map(int, input().split()))
k = int(input())
L = []
for i, v in enumerate(numl):
    if k - int(v) in numl:
        x = sorted([v, k - int(v)])
        if tuple(x) not in L:
            L.append(tuple(x))
print(sorted(L))
```

```python
#W6A15
pairl = input().split()
k = int(input())
d = {}
d2 = {}
for i in pairl:
    key, value = i.split(':')
    d[key] = value
    if int(value) > k:
        d2[key] = value
print(d2)
```
