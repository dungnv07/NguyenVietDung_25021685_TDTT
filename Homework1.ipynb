```python
ten = input('Ten chu ho: ')
kwh1 = int(input('Chi so thang truoc: '))
kwh2 = int(input('Chi so thang nay: '))
kwh = kwh2 - kwh1
price_section = [(50,1984),(50,2050),(100,2380),(100,2998),(100,3350),(float('inf'),3460)]
def total(kwh):
    total_price = 0
    for limit, priceper in price_section:
        if kwh > limit:
            total_price += limit * priceper
            kwh -= limit
        else:
            total_price += kwh * priceper
            break
    return round(total_price * 1.08)
print(f'Ho va ten: {ten}')
print(f'Tien phai tra la: {total(kwh)}')
```
