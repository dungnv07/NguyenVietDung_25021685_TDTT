```python
#W4A8
n = int(input())
def reverse(n):
    rev = 0
    while n >= 1:
        rev = rev * 10 + (n%10)
        n //= 10
    return rev
while n != reverse(n):
    n += reverse(n)
print(n)
```

```python
#W4A9
def diff(n):
    while n > 0:
        d = n % 10
        m = n // 10
        while m > 0:
            dm = m % 10
            if dm == d:
                return False
            m //= 10
        n //= 10
    return True
n = int(input())
if n < 0:
    print('no number')
sq = 0
for i in range(n+1):
    sq = i**2
    if diff(sq) == True:
        print(sq, end = ' ')
```

```python
#W4A10
def collatz(x): 
    chain = 1
    while x != 1:
        if x % 2 == 0:
            x /= 2
            chain += 1
        else:
            x = x*3 + 1
            chain += 1
    return chain
n = int(input())
L = 1
x = 0
for i in range(1, n+1):
    if collatz(i) > L:
        L = collatz(i)
        x = i
print(x, L)


    
        
```

```python
#W4A11
import math as m
def edivisor(a):
    even = 0
    for i in range(1, int(m.sqrt(a))+1):
        if a%i == 0:
            if i % 2 == 0:
                even += 1
            if (a/i) % 2 == 0 and a/i != i:
                even += 1
    return even
n = int(input())
print(edivisor(n))
```

```python
#W4A13
import math as m
def sumdiv(a):
    sum = 0
    for i in range(1, int(m.sqrt(a))+1):
        if a%i == 0 and a/i != i and i != 1:
            sum += i
            sum += a/i
        if i == 1 or a/i == i:
            sum += i
    return sum
a, b = map(int,input().split())
if sumdiv(a) == b or sumdiv(b) == a:
    print('True')
else:
    print('False')
print(sumdiv(a), sumdiv(b))
```
